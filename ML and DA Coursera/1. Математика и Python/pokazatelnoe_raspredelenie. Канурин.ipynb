```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
import math
```

# Показательное рарспределение
Было выбрано показательно распределение с параметром lambda=1\n
Дисперсия и мат.ожидание равны 1\n
Как видно из графиков ниже, плотность рспределения этого распределения значительно отличается от плотности нормального распределения

```python
grid = np.linspace(-2, 5, 100)
plt.plot(grid, stats.expon(scale=1).pdf(grid), color='red',label='Показательное распределение') 
plt.plot(grid, stats.norm(loc=1,scale=1).pdf(grid), color='blue',label='Нормальное распределение')
plt.legend(fontsize=10, loc=1)
plt.title("Графики плотностей различны распределений")
plt.show()
```

# Далее рассмотрим выборку из 1000 элементов

```python
viborka_pokazatelnoe=stats.expon(scale=1).rvs(size=1000)
grid = np.linspace(0, 8, 100)  
plt.hist(viborka_pokazatelnoe,bins=50,density=True,label='Нормированная гистограмма выборки') 
plt.plot(grid, stats.expon(scale=1).pdf(grid), color='red',label='Плотность случайной величины') 
plt.legend(fontsize=10, loc=1)
plt.title("График плотности показательного распредления и гистограмма выборки")
plt.show()
```

Теперь мы сравниваем распределение выборочного среднего с нормальным распределением для разных значений элементов выборки. В качестве параметров нормального распределения используем мат. ожидание и дисперсию деленную на число элементов выборки показательного распределения. (В нашем случае мат. ожидание=1, второй параметр=1/n)

```python
n=5
viborochnoe_srednee=np.zeros(1000)
for i in range(1000):
    viborka_pokazatelnoe1=stats.expon(scale=1).rvs(size=n)
    viborochnoe_srednee[i]=np.sum(viborka_pokazatelnoe1)/len(viborka_pokazatelnoe1)
plt.hist(viborochnoe_srednee,bins=50,density=True,label='Гистограмма распределения') 
grid = np.linspace(0, 3, 100)
plt.plot(grid, stats.norm(loc=stats.expon(scale=1).mean(), scale=stats.expon(scale=1).std()/math.sqrt(n)).pdf(grid), color='red',label='Плотность случайной величины')
print(stats.expon(scale=1).mean(),stats.expon(scale=1).std())
plt.legend(fontsize=8, loc=1)
plt.title("Гистограмма распределения выборочного среднего для 5 элемнтов в исходных выборках \n  в сравнении с нормальным распределнием")
plt.xlabel("Встречающиеся значения выборочного среднего")
plt.ylabel("Частота появления значений \n выборочного среднего")
plt.show()
```
